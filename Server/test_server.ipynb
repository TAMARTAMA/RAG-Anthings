```python
import json
from pathlib import Path
import requests

cfg = json.loads(Path("config.json").read_text(encoding="utf-8"))
host = cfg["server"].get("host", "127.0.0.1")
if host in {"0.0.0.0", "*"}:
    host = "127.0.0.1"
port = cfg["server"].get("port", 8002)
base_url = f"http://{host}:{port}"

payload = {"message": "How is the main server?"}
response = requests.post(f"{base_url}/ask", json=payload, timeout=60)
print("ASK status:", response.status_code)
print(response.json().get("answer", "").get("text", ""))
```

```python
import requests
import json

# כתובת השרת (תשני לפי הצורך)
SERVER_IP = "127.0.0.1"   # או localhost אם רץ מקומית
PORT = 8000

# כתובת מלאה
url = f"http://{SERVER_IP}:{PORT}/search"

# גוף הבקשה (JSON)
payload = {
    "query": ["machine learning", "AI", "neural networks"]
}

# שליחת הבקשה
try:
    response = requests.post(url, json=payload, timeout=30)

    # הדפסת הסטטוס
    print("Status:", response.status_code)

    # אם הצליח — הדפס את התוצאה בפורמט יפה
    if response.ok:
        print(json.dumps(response.json(), indent=2, ensure_ascii=False))
    else:
        print("Error:", response.text)

except requests.exceptions.RequestException as e:
    print("❌ שגיאה בשליחת הבקשה:", e)
```
